```python
online = 0
cash = 0
for i in range(0, 10087):
    if fine_df['cash_fine'][i] != '0':
        cash = cash+1
    else:
        online = online+1
        
print("Number of times an individual paid via cash:", cash)
print("Number of times an individual paid via online:", online)
```

```python
plot = ['Paid via cash', 'paid via online']
fig = go.Figure([go.Bar(x=plot, y=[cash, online])])
fig.show()
```

From this we can conclude most people prefer online fine payment that via cash payment

```python
ny_online= 0
ny_cash = 0
for i in range(0, 10087):
    if fine_df['state'][i] == "NY":
        #ny_cash += 1
        if fine_df['cash_fine'][i] != '0':
            ny_cash += 1
        else:
            ny_online += 1
            
            
ca_online= 0
ca_cash = 0
for i in range(0, 10087):
    if fine_df['state'][i] == "CA":
        if fine_df['cash_fine'][i] != '0':
            ca_cash += 1
        else:
            ca_online += 1


fl_online= 0
fl_cash = 0
for i in range(0, 10087):
    if fine_df['state'][i] == "FL":
        if fine_df['cash_fine'][i] != '0':
            fl_cash += 1
        else:
            fl_online += 1


tx_online= 0
tx_cash = 0
for i in range(0, 10087):
    if fine_df['state'][i] == "TX":
        if fine_df['cash_fine'][i] != '0':
            tx_cash += 1
        else:
            tx_online += 1

pa_online= 0
pa_cash = 0
for i in range(0, 10087):
    if fine_df['state'][i] == "PA":
        if fine_df['cash_fine'][i] != '0':
            pa_cash += 1
        else:
            pa_online += 1
```

```python
states = ['CA','NY','FL','TX','PA']
state_cash = []
state_cash.append(ca_cash)
state_cash.append(ny_cash)
state_cash.append(fl_cash)
state_cash.append(tx_cash)
state_cash.append(pa_cash)
state_cash
```

```python
state_onl = []
state_onl.append(ca_online)
state_onl.append(ny_online)
state_onl.append(fl_online)
state_onl.append(tx_online)
state_onl.append(pa_online)
state_onl
```

```python
states = ['CA','NY','FL','TX','PA']

fig = go.Figure(data=[
    go.Bar(name='Online', x=states, y=state_onl),
    go.Bar(name='Cash', x=states, y=state_cash)
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()
```
